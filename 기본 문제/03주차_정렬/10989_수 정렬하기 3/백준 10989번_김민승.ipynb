```python
##### 계수 정렬 알고리즘 #####
# 연속된 숫자가 아닌 1, 2, 3 그리고 갑자기 100이 나온다면 그 빈 공간에 해당하는 공간(메모리) 낭비가 생겨서 계수보다는 퀵이나 병합을 많이 사용함

N = int(input())
n = []

for i in range(N):
    a = int(input())
    n.append(a)

output = [0] * (N+1)                 # 계수 정렬을 위한 출력 배열 만들기. np.zeros를 쓸 수 없어서 이렇게 만듦! # 출력 배열은 입력값의 최대값보다 많아야 하므로 +1을 해줌

for i in range(N):                   # 지금은 range(N) = len(n) 이라 이렇게 쓰지만 여러 경우에도 적용하고 싶다면 range(len(n))을 쓰는 게 맞음
    output[n[i]] += 1                # 입력 배열과 동일한 출력 배열의 인덱스에 +1을 해줌. # 만약 [5, 4, 5, 3, 2, 1] 라고 한다면 출력 배열에는 [0, 1, 1, 1, 1, 2]로 입력 됨

for i in range(len(output)):         # i에 range(5) = 0~5를 for문으로 돌림
    for j in range(output[i]):       # 출력 배열을 불러와서 각 인덱스에 몇 개의 값이 들어가있는지 확인 및 아래에서 출력
        print(i)
```

```python
# 백준 제출용

import sys

N = int(sys.stdin.readline())

output = [0] * 10001

for i in range(N):
    output[int(sys.stdin.readline())] += 1

for i in range(len(output)):
    for j in range(output[i]):
        print(i)
```
