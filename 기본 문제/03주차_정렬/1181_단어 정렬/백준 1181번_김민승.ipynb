```python
N = int(input())                       # 총 들어갈 단어 개수
st = []

for i in range(N):
    word = input().split()[0]          # 단어 입력 받기 # split().[0]을 해주는 이유가 뭔지 아직도 잘 모르겠다ㅠ
    st.append(word)                    # 빈 리스트에 추가해서 단어로 이루어진 리스트 생성

lst = list(set(st))                    # set으로 중복 단어 제거
lst.sort(key=lambda x: (len(x), x))    # lambda로 다중 정렬. 첫번째 조건은 스펠링 수대로, 두번째 조건은 사전 순대로

for i in lst:
    print(i)
```

```python
#### 백준 제출용 ####

import sys

N = int(sys.stdin.readline())
st = []

for i in range(N):
    word = sys.stdin.readline().split()[0]
    st.append(word)

st = list(set(st))
st.sort(key=lambda x: (len(x), x))

for i in st:
    print(i)
```
