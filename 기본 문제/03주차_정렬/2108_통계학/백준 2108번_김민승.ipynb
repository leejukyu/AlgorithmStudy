```python
from collections import Counter

N = int(input())
n = []

for i in range(N):
    num = int(input())
    n.append(num)


def mean(lst):
    return round(sum(lst)/len(n))                    # 조건을 보고 round((sum(lst)/len(n)), 0)으로 썼으나 값이 2.0으로 나오는 것을 보고 수정함

def center(lst):
    lst.sort()                                       # 오름차순으로 정렬하라는 조건이 있어서 삽입 ## 이때 적용한 sort()는 아래에도 영향을 주는 것처럼 보임, 왜 그런 걸까?
    middle = lst[len(lst)//2]                        # 원래 len(n) // 2로 나오는 값을 바로 반환시켰으나 n=[4000]인 예제에서 0이 출력이 되길래 인덱스 값으로 출력하기로 함
    return middle

def mode(lst):                                       # 보다 쉽게 값을 구하기 위해 counter 사용
    num = Counter(lst)
    nums = num.most_common()

    if len(lst) > 1 and nums[0][1] == nums[1][1]:    # 만약 주어진 리스트 원소가 1개보다 많고 같은 최빈값을 가지는 원소가 있다면
        return nums[1][0]                            # 그 중 2번째에 있는 값을 내뱉는다 (백준에 나와있는 조건)
    else:
        return nums[0][0]

def scope(lst):
    return max(lst) - min(lst)

print(mean(n))
print(center(n))
print(mode(n))
print(scope(n))
```

```python
##### 백준 제출용 #####

import sys 
from collections import Counter

N = int(sys.stdin.readline())
n = []

for i in range(N):
    num = int(sys.stdin.readline())
    n.append(num)


def mean(lst):
    return round(sum(lst)/len(n))

def center(lst):
    lst.sort()
    middle = lst[len(lst)//2]
    return middle

def mode(lst):
    num = Counter(lst)
    nums = num.most_common()

    if len(lst) > 1 and nums[0][1] == nums[1][1]:
        return nums[1][0]
    else:
        return nums[0][0]

def scope(lst):
    return max(lst) - min(lst)

print(mean(n))
print(center(n))
print(mode(n))
print(scope(n))
```
