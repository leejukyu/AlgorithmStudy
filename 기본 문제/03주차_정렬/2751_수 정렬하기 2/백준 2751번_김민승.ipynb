```python
##### 병합 정렬 알고리즘1 #####
# 백준 제출용

def merge(array):
    if len(array) < 2:                               # 입력 리스트 안의 갯수가 2개 미만이면 배열을 그대로 출력한다 # 종료문
        return array

    benchmark = len(array) // 2                      # 입력 리스트를 2로 나눠서 기준점을 잡는다
    group1 = merge(array[:benchmark])                # 기준점을 토대로 2개의 리스트로 나눈다
    group2 = merge(array[benchmark:])                # 이때 재귀함수를 써서 각 리스트의 수를 정렬시킨다

    result = []

    g1 = g2 = 0

    while g1 < len(group1) and g2 < len(group2):     # 1그룹이나 2그룹이 0보다 크지 않을 경우 while문을 종료한다
        if group1[g1] < group2[g2]:                  # 1그룹의 첫번째 인덱스의 수가 2그룹의 첫번째 인덱스의 수보다 작으면
            result.append(group1[g1])                # result 리스트에 1그룹의 수를 추가한다 
            g1 += 1                                  # 이후 1그룹은 다음 인덱스로 넘어간다
        else:
            result.append(group2[g2])                # 위를 만족하지 못할 경우 2그룹의 첫번째 인덱스 수를 result 리스트에 추가한다
            g2 += 1                                  # 이후 2그룹은 다음 인덱스로 넘어간다
    result += group1[g1:]
    result += group2[g2:]                            # 위를 반복해서 넣고 남은 수를 result에 넣는다
    return result

N = int(input())
n = []

for i in range(N):
    a = int(input())
    n.append(a)

for i in merge(n):
    print(i)

# import sys

# N = int(sys.stdin.readline())
# n = []

# for i in range(N):
#     a = int(sys.stdin.readline())
#     n.append(a)
```

```python
##### 병합 정렬 알고리즘2 #####

def merge(array):
    n = len(array)

    if n <= 1:
        return array
    
    benchmark = n // 2
    
    group1 = merge(array[:benchmark])
    group2 = merge(array[benchmark:])

    result = []

    while group1 and group2:
        if group1[0] < group2[0]:
            result.append(group1.pop(0))         # 위의 조건을 만족했을 때 1그룹에서 첫번째 수를 result에 추가함과 동시에 1그룹에서 첫번째 수 삭제
        else:
            result.append(group2.pop(0))
    while group1:
        result.append(group1.pop(0))
    while group2:
        result.append(group2.pop(0))
    return result

N = int(input())
n = []

for i in range(N):
    a = int(input())
    n.append(a)

for i in merge(n):
    print(i)
```
